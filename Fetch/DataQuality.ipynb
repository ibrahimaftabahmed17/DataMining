```python
users_array = []
with open('users.json') as f:
    for jsonObj in f:
        userDict = json.loads(jsonObj)
        users_array.append(userDict)

f.close()
```

```python
df = pd.DataFrame()
for i in range(0,len(users_array)):
    df = df.append(pd.json_normalize(users_array[i]))
```
